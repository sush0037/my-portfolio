```python
text_1 = "We do not like overall experience with the product"
```

Using transformers for text classification into positive or Negative

```python
from transformers import pipeline
classifier = pipeline('text-classification')
import pandas as pd

output_1 = classifier(text_1)
pd.DataFrame(output_1)
```

```python
from transformers import set_seed
set_seed(42)  # Set the seed to get reproducible results
generator = pipeline('text-generation')
response = "Dear valued customer, thanks for writing in! I am sorry hear your experience "
prompt = text_1+'\n\Customer Service Response:\n'+response
output_1 = generator(prompt, max_length=200)
```

```python
print(output_1[0]['generated_text'])
```

```python
from transformers import pipeline

# Initialize a translation pipeline (e.g., English to French)
# You can find models for other language pairs on Hugging Face Model Hub (e.g., opus-mt-en-de for English to German)
translator = pipeline("translation_en_to_mr", model="Helsinki-NLP/opus-mt-en-mr")

# Text to translate
text_to_translate = "Hello, how are you today? "

# Perform translation
translated_output = translator(text_to_translate)

# Print the translated text
print(f"Original English: {text_to_translate}")
print(f"Translated marathi: {translated_output[0]['translation_text']}")
```
